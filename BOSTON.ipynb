```python
df = pd.read_csv('https://raw.githubusercontent.com/Prago2001/LP-5/master/DL/Assignment_1/Boston.csv')
```

```python
df.head(10)
```

```python
df.drop(['Unnamed: 15', 'Unnamed: 16'], axis=1, inplace=True)
```

```python
df.drop('CAT. MEDV', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()['MEDV'].sort_values()
```

```python
X = df.loc[:,['LSTAT', 'PTRATIO', 'RM']]
Y = df.loc[:,['MEDV']]
X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(128, input_shape=(3,), activation = 'relu', name='input'))
model.add(Dense(64, activation='relu', name='layer_1'))
model.add(Dense(1, activation='linear', name='output'))
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()
```

```python
model.fit(x_train, y_train, epochs=100, validation_split=0.05)
```

```python
output = model.evaluate(x_test, y_test)
```

```python
print(f"Mean Squared Error: {output[0]}"
      ,f"Mean Absolute Error: {output[1]}",sep="\n")
```

```python
y_pred = model.predict(x=x_test)
```
