```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

traindf = pd.read_csv('/content/fashion-mnist_train.csv')
```

```python
testdf=pd.read_csv('/content/fashion-mnist_test.csv')
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat','Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
x_train = traindf.iloc[:,1:].to_numpy()
x_train = x_train.reshape([-1,28,28,1])
x_train=x_train/255
```

```python
y_train = traindf.iloc[:,0].to_numpy()
```

```python
x_test=testdf.iloc[:,1:].to_numpy()
x_test=x_test.reshape([-1,28,28,1])
x_test = x_test/255
```

```python
y_test=testdf.iloc[:,0].to_numpy()
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D, Dropout
```

```python
model = Sequential()
```

```python
model.add(Conv2D(filters=64,kernel_size=(3,3),input_shape=(28,28,1),activation='relu'))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(rate=0.3))
model.add(Flatten())
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=10, activation='sigmoid'))
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train,y_train,epochs=50,batch_size=1200,validation_split=0.05)
```

```python
evaluation = model.evaluate(x_test,y_test)
```

```python
evaluation[1]
```

```python
y_probas = model.predict(x_test)
```

```python
y_pred = y_probas.argmax(axis=-1)
```

```python
y_pred
```

```python
plt.figure(figsize=(10,10),)
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_test[i], cmap=plt.cm.binary)
#     plt.xlabel(f"True Class:{y_test[i]}")
    plt.title(f"Pred:{class_names[y_pred[i]]}")
plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
num_classes = 10
class_names = ["class {}".format(i) for i in range(num_classes)]
cr = classification_report(y_test, y_pred, target_names=class_names)
print(cr)
```
